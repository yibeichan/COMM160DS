# Introduction to Data Visualization with Matplotlib and Advanced Visualization with Seaborn

## Part 1: Introduction to Matplotlib

### 1.1: Matplotlib Overview
Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides an object-oriented API for embedding plots into applications and a scripting layer for creating simple plots quickly.

To use Matplotlib, we need to import it. The most common way to import Matplotlib is as follows:

```python
import matplotlib.pyplot as plt
```

### 1.2: Basic Plotting with Matplotlib
In this section, we will cover four basic types of plots: line plots, scatter plots, bar plots, and histograms.

#### Line Plot

A line plot is a way to display data along a number line. To create a line plot, use the `plot()` function:

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.show()
```

#### Scatter Plot

A scatter plot uses dots to represent values for two different numeric variables. To create a scatter plot, use the `scatter()` function:

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y)
plt.show()
```

#### Bar Plot

A bar plot represents data using rectangular bars. To create a bar plot, use the `bar()` function:

```python
x = ['Category 1', 'Category 2', 'Category 3']
y = [10, 20, 30]

plt.bar(x, y)
plt.show()
```

#### Histogram

A histogram is a graphical representation of the distribution of a dataset. To create a histogram, use the `hist()` function:

```python
data = [1, 1, 1, 2, 2, 3, 4, 4, 4, 4]

plt.hist(data, bins=4)
plt.show()
```

### 1.3: Customizing Plots

In this section, we will cover customizing plots by changing colors, markers, line styles, titles, labels, legends, axis limits, and ticks.

#### Colors, Markers, and Line Styles

Customize the appearance of your plot using various options in the plot() function:

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color='orange', marker='o', linestyle='--', linewidth=3, markersize=10)
plt.show()
```

#### Titles, Labels, and Legends

Add a title, x-axis label, and y-axis label to your plot using the `title()`, `xlabel()`, and `ylabel()` functions:

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.title('Line Plot Example')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.show()
```

#### Axis Limits and Ticks

Control the limits and ticks of the `x` and `y` axes using the `xlim()`, `ylim()`, `xticks()`, and `yticks()` functions:

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 1000]

plt.scatter(x, y)
plt.title('Line Plot Example')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.xlim(0, 6)
plt.ylim(0, 17)
# plt.xticks([0, 1, 2, 3, 4, 5, 6])
# plt.yticks([0, 5, 10, 15, 20, 25, 30])
plt.show()
```

## Part 2: Advanced Visualization with Seaborn

### 2.1: Seaborn Overview
Seaborn is a statistical data visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. To use Seaborn, we need to import it:

```python
import seaborn as sns
```

### 2.2: Basic Plotting with Seaborn
In this section, we will cover four basic types of plots: distplot, boxplot, violin plot, and pair plot.

#### Distplot

A distplot combines a histogram and a kernel density estimate (KDE) plot. To create a distplot, use the `distplot()` function:

```python
import numpy as np

data = np.random.normal(size=100)

sns.displot(data)
plt.show()
```

```python
sns.histplot(data)
plt.show()
```

#### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five-number summary ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum"). To create a boxplot, use the `boxplot()` function:

```python
import pandas as pd
```

```python
df = pd.read_csv("pre-course_survey.csv")
df
```

```python
df.columns
```

```python
sns.boxplot(x='4. Are you familiar with Jupyter Notebooks or JupyterHub?', y='1. On a scale of 1 to 5, how would you rate your current knowledge of Python programming?', data=df)
plt.xlabel('Knowledge of Jupyter')
plt.ylabel('Knowledge of Python')
plt.show()
```

```python
data = sns.load_dataset('iris')
sns.boxplot(x='species', y='sepal_length', data=data)
plt.show()
```

```python
type(data)
```

#### Violin Plot

A violin plot is a combination of a boxplot and a kernel density plot, showing the distribution of the data across different categories. To create a violin plot, use the `violinplot()` function:

```python
data = sns.load_dataset('iris')
sns.violinplot(x='species', y='sepal_length', data=data)
plt.show()
```

```python
data
```

#### Pair Plot

A pair plot is a grid of scatterplots that display the relationship between pairs of variables in a dataset. To create a pair plot, use the `pairplot()` function:

```python
data = sns.load_dataset('iris')
sns.pairplot(data, hue='species')
plt.show()
```

### 2.3: Customizing Seaborn Plots
In this section, we will cover customizing plots by changing colors, markers, line styles, titles, labels, legends, axis limits, and ticks.

#### Colors and Styles

Seaborn provides various color palettes and styles to make your plots visually appealing. Set the color palette and style using the `set_palette()` and `set_style()` functions:

```python
sns.set_palette('pastel')
sns.set_style('whitegrid')

data = sns.load_dataset('iris')
sns.boxplot(x='species', y='sepal_length', data=data)
plt.show()
```

#### Titles, Labels, and Legends

Add a title, x-axis label, and y-axis label to your plot using the same Matplotlib functions:

```python
data = sns.load_dataset('iris')
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Boxplot Example')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

#### Axis Limits and Ticks

Control the limits and ticks of the x and y axes using the same Matplotlib functions:

```python
data = sns.load_dataset('iris')
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Boxplot Example')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
# plt.xlim(-0.5, 2.5)
plt.ylim(3, 9)
plt.xticks(range(3), ['Setosa', 'Versicolor', 'Virginica'])
plt.yticks(range(3, 10))
plt.show()
```
